```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as pdr
```

```python
# 台積電、Yahoo，2019-01-01~2020-12-31
data = pdr.DataReader("2330.TW", "yahoo", "2019-01-01", "2020-12-31")
```

```python
# 收盤價
close_price = data["Close"]
```

```python
# 收盤價
close_price.plot(label="Close Price")
# 20日移動平均線
close_price.rolling(window = 20).mean().plot(label = "20MA")
# 60日移動平均線
close_price.rolling(window = 60).mean().plot(label = "60MA")
# 圖例標示
plt.legend(loc = "best")
# Show
plt.show()
```
